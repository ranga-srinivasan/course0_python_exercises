<b>4. Write a Python program to create a 1x2 grid of subplots. The first subplot should show a simple line plot with data [1, 2, 3, 4] and [10, 20, 30, 40]. The second subplot should display a bar plot with the same x-axis values and corresponding y-axis values [5, 10, 15, 20].</b>

<b>Answer:</b>

<b>5. Write a Python program to find the maximum number in a list of numbers using a built-in function.<br/>
Sample Input -> [10, 4, 45, 99, 12]<br/>
Sample Output -> The maximum value is: 99</b>

<b>Answer:</b>

```python
def find_max(list):
    max = list[0]
    for i in list:
        if i > max:
            max = i
    return max

list = [10, 4, 45, 99, 12]
print('The maximum value is: ', find_max(list))
```

<b>6. Create a Python program that defines a custom function to calculate the factorial of a number. <br/>The function should take one integer as an argument and return its factorial.
Sample Input -> 5<br/>
Sample Output -> The factorial of 5 is: 120</b>

<b>Answer:</b>

```python
def factorial(n):
    if n == 0 or n == 1:
        return n
    return n*factorial(n-1)

print('factorial of 5 is: ', factorial(5))
```

<b>8. Write a Python program that checks if a file exists before performing operations like reading or writing.</b>

<b>Answer:</b>

```python
import os

def read_content(filename):
    f = open(filename, 'r')
    content = f.read()
    f.close()
    return content

def check_and_open_file(filename):
    if os.path.exists(filename):
        print(filename, ' exists.', ' Proceeding further to read the content')
        content = read_content(filename)
        print(content)
    else:
        print('ERROR: ', filename, ' does not exist!')
    
check_and_open_file('sample_file.txt')
check_and_open_file('unknown.txt')
```

<b>9. Write a program to solve the below equations<br/>
• 2x+y=5<br/>
• x−y=1</b>

<b>Answer:</b>

```python
A = np.array([
                [2, 1], 
                [1, -1]])
B = np.array([
                5,
                1])

X = np.linalg.solve(A, B)
print('Solution of the linear equations: ')
print('x = ', X[0])
print('y = ', X[1])
```

<b>10. Write a program to integrate the function f(x)=x2 from 0 to 1.</b>

<b>Answer:</b>

```python
def f(x):
    return x*x

x = np.linspace(0, 1, 1000)
y = f(x)
integral = np.trapz(y, x)
print('Integral of f(x) = x^2 from 0 to 1 is: ', integral)
```

<b>11. Write a program to find the inverse by inv() function and determinant det() function.</b>

<b>Answer:</b>

```python
def find_det_and_inv_of_matrix(A):
    detA = np.linalg.det(A)
    print('Determinant of the matrix A is: ', detA)
    if detA != 0:
        invA = np.linalg.inv(A)
        print('Inverse of the matrix A is: ', invA)
    else:
        print('Matrix A is not invertible!')

A1 = np.array([[2, 1],
               [1, -1]])

find_det_and_inv_of_matrix(A1)
print()
print()
A2 = np.array([[2, 1],
               [-2, -1]])

find_det_and_inv_of_matrix(A2)
```

<b>13. Write a program to perform Arithmetic Operations on Multi-Dimensional Arrays.</b>

<b>Answer:</b>

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print('Matrix A:')
print(A)

print()

print('Matrix B:')
print(B)

print()

print('Matrix Addition:')
print(A + B)

print()

print('Matrix Subtraction:')
print(A - B)

print()

print('Matrix Multiplication (element-wise):')
print(A * B)

print()

print('Matrix Division (element-wise):')
print(A / B)
```

<b>14. Create a python program to check the missing values in a dataset. </b>

<b>Answer:</b>

```python
courses = {
    "name": ["Linear Algebra", "Calculus", "Probability", None],
    "course_number": ['LINAL', None, 'CALCI', 'PROBS'],
    "instructor": ["Dr M.Radhakrishnan", "Dr P.K.Raman", None, "Dr Shivprasad Yadav"]
}

df = pd.DataFrame(courses)

print("Dataset:")
print(df)

print()

print("Missing Values Summary:")
print(df.isnull().sum())

print()

print("Any missing values in the dataset?")
print(df.isnull().values.any())

print()

print("Total number of missing values in the dataset:", df.isnull().sum().sum())
```

<b>15. Write a python program to group and aggregate the data.</b>

<b>Answer:</b>

```python
student_profiles = {
    "course": ["Linear Algebra", "Advanced Calculus", "Linear Algebra", "Advanced Calculus", "Physics", "Physics", "Linear Algebra"],
    "name": ["Ram", "Krishnan", "Swamy", "Anand", "Balaji", "Ramesh", "Prasad"],
    "marks": [65, 72, 61, 70, 45, 48, 52],
    "CGPA": [7.2, 5.5, 6.7, 9.0, 8.5, 7.8, 8.3]
}

df = pd.DataFrame(student_profiles)

print("Dataset:")
print(df)

grouped = df.groupby("course").agg({
    "marks": ["sum", "mean"],
    "CGPA": ["max", "min"]
})

print("\nGrouped and Aggregated Data:")
print(grouped)
```
